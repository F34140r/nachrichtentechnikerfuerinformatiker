```python
import scipy.fftpack
```

```python
def fun(t):
    mu = 0.5
    sigma = 0.1 ** 0.5
    freq = 100
    x = (1+np.sin(2*math.pi*freq*t)) * np.exp(-((t-mu) / sigma)**2)
    return x
```

```python
ts = np.linspace(0,1,500)
dt = np.diff(ts).mean()
xs = fun(ts)
plt.plot(ts,xs)
```

```python
spectrum = np.fft.fft(xs)
fftfreq = np.fft.fftfreq(ts.shape[-1],dt)
fftfreq = fftfreq[:fftfreq.shape[0]//2]
spectrum = spectrum[:spectrum.shape[0]//2]
```

```python
#plt.figure(figsize=(15,10))
plt.plot(fftfreq, np.abs(spectrum))
```
